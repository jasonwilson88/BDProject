```python
from urllib.request import urlopen
from PIL import Image
import os.path
import random
```

```python
if os.path.isfile("photos/sla1.jpg"):
    text_file = []
else:
    text_file = open("urls.txt", "r")
```

```python
for index, url in enumerate(text_file):
    print(url)
    try:
        link = urlopen(url)
        name = "photos/sla%s.jpg" % (index+1)
        with open(name, "wb") as output:
            output.write(link.read())
    except:
        print("Unable to create %s" % name)
        continue
```

```python
path = "photos/"
loopCount = 0
while loopCount < 10:
    random_filename = random.choice([
        x for x in os.listdir(path)
        if os.path.isfile(os.path.join(path, x))
    ])
 
    print(random_filename)
    image = Image.open("/home/jack/BDProject/Part2/six/photos/" + str(random_filename))
    image.show()
    loopCount++
```
